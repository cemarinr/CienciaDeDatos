## **1. Descargar el dataset**
El primer paso consiste en leer el archivo CSV que contiene el dataset. Para ello, utilizamos la biblioteca `pandas` que permite cargar datos desde una URL de un archivo CSV.

El archivo se encuentra almacenado en un repositorio de GitHub. Usamos la función `pd.read_csv()` de `pandas`, que permite cargar el contenido del archivo CSV en un DataFrame, lo cual facilita su manipulación para el análisis.

## Hay 14 atributos en cada caso del conjunto de datos. Estos son:
- **CRIM**: tasa de criminalidad per cápita por ciudad.
- **ZN**: proporción de terreno residencial zonificado para lotes de más de 25,000 pies cuadrados.
- **INDUS**: proporción de acres comerciales no minoristas por ciudad.
- **CHAS**: variable ficticia del río Charles (1 si el tramo limita con el río; 0 en caso contrario).
- **NOX**: concentración de óxidos de nitrógeno (partes por cada 10 millones).
- **RM**: número promedio de habitaciones por vivienda.
- **AGE**: proporción de unidades ocupadas por sus propietarios construidas antes de 1940.
- **DIS**: distancias ponderadas a cinco centros de empleo de Boston.
- **RAD**: índice de accesibilidad a carreteras radiales.
- **TAX**: tasa de impuesto sobre la propiedad de valor total por cada $10,000.
- **PTRATIO**: proporción de alumnos por maestro por ciudad.
- **B**: 1000(Bk - 0.63)^2 donde Bk es la proporción de personas de raza negra por ciudad.
- **LSTAT**: % de la población de estatus socioeconómico bajo.
- **MEDV**: Valor mediano de las viviendas ocupadas por sus propietarios en miles de dólares.


```python
# Importamos las bibliotecas necesarias: 'pandas' para manipulación de datos y 'numpy' para cálculos matemáticos
import pandas as pd
import numpy as np

# URL del archivo CSV ubicado en un repositorio de GitHub
# En este paso, utilizamos la función pd.read_csv() para leer el archivo y cargarlo en un DataFrame de pandas
url = 'https://raw.githubusercontent.com/selva86/datasets/refs/heads/master/BostonHousing.csv'
# Leemos el archivo CSV y lo cargamos en un DataFrame llamado 'df' para trabajar con los datos en forma tabular
df = pd.read_csv(url)
# Visualizamos las primeras filas del DataFrame para asegurarnos de que los datos se han cargado correctamente
df.head()
```

## **2. Cálculo de estadísticas descriptivas**

En este paso, se calcula la media, mediana y desviación estándar de las columnas numéricas del dataset. Estas métricas nos ayudarán a obtener una visión general de la distribución de los datos y su dispersión.

- **Media**: Es el promedio de los valores de una columna.
- **Mediana**: Es el valor central que separa la mitad superior de la mitad inferior de los datos.
- **Desviación estándar**: Mide la dispersión de los datos respecto a la media. Un valor alto indica que los datos están más dispersos, mientras que un valor bajo sugiere que están concentrados alrededor de la media.

El siguiente código realiza el cálculo de estas métricas:




```python
import numpy as np

# Cálculo de la media
# Utilizamos la función mean() para obtener el valor promedio de cada columna
media = df.mean()

# Cálculo de la mediana
# La función median() se usa para encontrar el valor que divide los datos en dos partes iguales
mediana = df.median()

# Cálculo de la desviación estándar
# La función std() mide qué tanto se desvían los datos respecto a la media
desviacion_estandar = df.std()

# Imprimimos las estadísticas calculadas para verificar los resultados
print("Media:\n", media)
print("Mediana:\n", mediana)
print("Desviación Estándar:\n", desviacion_estandar)
```

## **2. Cálculo de la Media**

En este paso, calculamos la **media** para cada columna numérica del dataset. La media representa el promedio de los valores de una columna y nos da una idea general del valor típico en los datos.

El código siguiente realiza el cálculo de la media y organiza los resultados en una tabla:

```python
import numpy as np
import matplotlib.pyplot as plt

# Cálculo de la media usando NumPy para cada columna numérica
media = np.mean(df, axis=0)

# Crear un DataFrame para las medias utilizando pandas
df_media = pd.DataFrame(media, index=df.columns, columns=['Media'])

# Mostrar la tabla de medias
print(df_media)

# Generar un gráfico de barras para visualizar la media de cada columna
plt.figure(figsize=(10, 6))
df_media['Media'].plot(kind='bar', color='skyblue')
plt.title('Media de las Columnas')
plt.ylabel('Valor Medio')
plt.xlabel('Columnas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## **3. Cálculo de la Mediana con NumPy y Gráfico**

En este paso, calculamos la **mediana** de las columnas numéricas del dataset. La mediana es el valor central en una distribución. Después de calcular la mediana, organizamos los resultados en una tabla y generamos un gráfico de barras para visualizar la mediana de cada columna.

```python
# Cálculo de la mediana usando NumPy para cada columna numérica
mediana = np.median(df, axis=0)

# Crear un DataFrame para las medianas utilizando pandas
df_mediana = pd.DataFrame(mediana, index=df.columns, columns=['Mediana'])

# Mostrar la tabla de medianas
print(df_mediana)

# Generar un gráfico de barras para visualizar la mediana de cada columna
plt.figure(figsize=(10, 6))
df_mediana['Mediana'].plot(kind='bar', color='lightgreen')
plt.title('Mediana de las Columnas')
plt.ylabel('Valor de la Mediana')
plt.xlabel('Columnas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## **4. Cálculo de la Desviación Estándar con NumPy y Gráfico**

Finalmente, calculamos la **desviación estándar** de cada columna numérica. La desviación estándar mide la variabilidad de los datos. Después de calcularla, organizamos los resultados en una tabla y generamos un gráfico para visualizar la dispersión de los datos.

```python
desviacion_estandar = np.std(df, axis=0)

# Crear un DataFrame para las desviaciones estándar utilizando pandas
df_desviacion = pd.DataFrame(desviacion_estandar, index=df.columns, columns=['Desviación Estándar'])

# Mostrar la tabla de desviación estándar
print(df_desviacion)

# Generar un gráfico de barras para visualizar la desviación estándar de cada columna
plt.figure(figsize=(10, 6))
df_desviacion['Desviación Estándar'].plot(kind='bar', color='salmon')
plt.title('Desviación Estándar de las Columnas')
plt.ylabel('Valor de la Desviación Estándar')
plt.xlabel('Columnas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
